```python
def split_image_into_channels(img):
    red_channel = img[:, :, 0]
    green_channel = img[:, :, 1]
    blue_channel = img[:, :, 2]
    return red_channel, green_channel, blue_channel
```

```python
def merge_channels(red_channel, green_channel, blue_channel):
    return np.stack([red_channel, green_channel, blue_channel], axis=2)
```

```python
def channel_adjust(channel, values):
    orig_size = channel.shape
    flat_channel = channel.flatten()
    adjusted = np.interp(
        flat_channel,
        np.linspace(0, 1, len(values)),
        values)

    return adjusted.reshape(orig_size)
```

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img = mpimg.imread("Q6.png")
r, g, b = split_image_into_channels(img)
plt.imshow(img)
```

```python
#Valencia
r_boost_lower = channel_adjust(r, [
    0, 0.199, 0.334, 0.570, 0.895, 1])
g_boost_lower = channel_adjust(g, [
    0, 0.071, 0.238, 0.408, 0.581, 0.832, 1])
b_boost_lower = channel_adjust(b, [
    0, 0.167, 0.314, 0.487, 0.667, 0.864, 1])
```

```python
Valencia = merge_channels(r_boost_lower, g_boost_lower, b_boost_lower)
plt.imshow(Valencia)
```

```python
#Rise
r_boost_lower = channel_adjust(r, [
    0, 0.120, 0.510, 0.667, 0.915, 1])
g_boost_lower = channel_adjust(g, [
    0, 0.120, 0.255, 0.393, 0.597, 0.824, 1])
b_boost_lower = channel_adjust(b, [
    0, 0.160, 0.322, 0.471, 0.687, 0.863, 1])
```
